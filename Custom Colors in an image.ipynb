##Creating shades of Red- Dense

```python
img=[]
for i in range(256):
    t=[]
    for j in range(256):
        t.append([i,0,0])
 
    img.append(t)
img=np.array(img)  
plt.imshow(img)

```

Creating shades of Green- Dense

```python
img=[]
for i in range(256):
    t=[]
    for j in range(256):
        t.append([0,i,0])
 
    img.append(t)
img=np.array(img)  
plt.imshow(img)

```

Creating shades of blue-Dense

```python
img=[]
for i in range(256):
    t=[]
    for j in range(256):
        t.append([0,0,i])
 
    img.append(t)
img=np.array(img)  
plt.imshow(img)

```

#Merging these three shades

```python
img=[]
for i in range(256):
    t=[]
    for j in range(1536):         # red to black
        t.append([i,0,0])
 
    img.append(t)
for i in range(256,0,-1):
    t=[]                              #black to Red
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)



for i in range(256):
    t=[]
    for j in range(1536):        # green to black
        t.append([0,i,0])
 
    img.append(t)
for i in range(256,0,-1):          #from black to green 
    t=[]
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

for i in range(256):
    t=[]
    for j in range(1536):        # blue to black
        t.append([0,0,i])
 
    img.append(t)
for i in range(256,0,-1):          #from black to blue 
    t=[]
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)

img=np.array(img)
plt.imshow(img)
```
