```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
## [R  , G  , B]
## [255 , 0 , 0]-Red
## [0 , 255 , 0]-Green
## [ 0 , 0 , 255 ]-Blue
```

1.Creating Red Images


```python
arr=np.array([[[255,0,0],[255,0,0],[255,0,0]],
              [[255,0,0],[255,0,0],[255,0,0]],
              [[255,0,0],[255,0,0],[255,0,0]]])
plt.imshow(arr)
```

```python
arr=np.array([[[0,255,0],[0,255,0],[0,255,0]],
              [[0,255,0],[0,255,0],[0,255,0]],
              [[0,255,0],[0,255,0],[0,255,0]]])
plt.imshow(arr)
```
